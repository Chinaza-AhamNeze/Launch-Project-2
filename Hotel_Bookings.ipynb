## Creating a Nested Reservations Record

```python
nested_reservations.delete_many({})
```

```python
reservations_guests = Reservations.merge(
    guests, how="left", left_on="guest_id", right_on="Guest_ID"
)
```

```python
reservations_guests_rooms = reservations_guests.merge(
    Rooms, how="left", on="room_id"
)
```

```python
nested_df = reservations_guests_rooms.merge(
    Payments, how="left", on="reservation_id"
)
```

```python
def create_nested_record(row):
    return {
        "reservation_id": row["reservation_id"],
        "check_in_date": str(row["check_in_date"]),
        "check_out_date": str(row["check_out_date"]),
        "is_occupied": row["is_occupied"],
        "cleanliness_state": row["cleanliness_state"],
        "guest": {
            "guest_id": row["Guest_ID"],
            "first_name": row["First_Name"],
            "last_name": row["Last_Name"],
            "email": row["Email"],
            "phone": row["Phone number"],
            "membership_status": row["Membership Status"],
            "address": row["Address"]
        },
        "room": {
            "room_id": row["room_id"],
            "number": row["number"],
            "view_type": row["view_type"],
            "capacity": row["capacity"],
            "room_type": row["room_type"],
            "availability": row["availability"],
            "last_maintenance_date": str(row["last_maintenance_date"]),
            "accessibility_accomodation": row["accessibility_accomodation"]
        },
        "payment": {
            "payment_id": row["payment_id"],
            "payment_method": row["payment_method"],
            "payment_status": row["payment_status"],
            "amount": row["amount_in_$"],
            "payment_date": str(row["payment_dates"])
        }
    }
```

```python
nested_records = nested_df.apply(create_nested_record, axis=1).tolist()
```

```python
nested_reservations = db.Nested_data
```

```python
nested_reservations.insert_many(nested_records)
```

## Queries

### Query to find all guests who have spent more than $500

```python
pipeline = [
    {
        "$group": {
            "_id": "$guest.email",
            "totalSpent": { "$sum": "$payment.amount" },
            "firstName": { "$first": "$guest.first_name" },
            "lastName": { "$first": "$guest.last_name" }
        }
    },
    {
        "$addFields": {
            "fullName": { "$concat": ["$firstName", " ", "$lastName"] }
        }
    },
    {
        "$match": {
            "totalSpent": { "$gt": 500 }
        }
    },
    {
        "$sort": { "totalSpent": -1 }
    }
]
```

```python
result = list(nested_reservations.aggregate(pipeline))
for doc in result:
    print(doc)
```

### Revenue By Membership

```python
pipeline = [
    {
        "$group": {
            "_id": {
                "membership": "$guest.membership_status",
                "roomType": "$room.room_type"
            },
            "totalRevenue": { "$sum": "$payment.amount" },
            "reservationCount": { "$sum": 1 }
        }
    },
    {
        "$sort": {
            "_id.membership": 1,           # Sort by membership (alphabetically)
            "totalRevenue": -1             # Then by revenue descending
        }
    }
]
```

```python
result = list(nested_reservations.aggregate(pipeline))
for doc in result:
    print(doc)
```

### Most booked viewtypes

```python
pipeline = [
    {
        "$group": {
            "_id": "$room.number",
            "Membership": { "$first": "$guest.membership_status" },
            "View": { "$first": "$room.view_type" },
            "count": { "$sum": 1 }
        }
    },
    { "$match": { "count": { "$gt": 3 } } }
]
```

```python
result = list(nested_reservations.aggregate(pipeline))
for doc in result:
    print(doc)
```

#### We can infer that customers prefer Ocean/Pool side views

### Top-Spending Guests with their contact info

```python
pipeline = [
    {
        "$group": {
            "_id": "$guest.email",
            "name": { "$first": { "$concat": [{ "$toString": "$guest.first_name" },
                        " ",
                        { "$toString": "$guest.last_name" }] } },
            "phone": { "$first": "$guest.phone" },
            "totalSpent": { "$sum": "$payment.amount" }
        }
    },
    { "$sort": { "totalSpent": -1 } },
    { "$limit": 10 }
]
```

```python
result = list(nested_reservations.aggregate(pipeline))
for doc in result:
    print(doc)
```
